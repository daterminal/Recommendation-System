1차 고민:
weight는 여기서 각 패턴의 중요도인데, 이 weight를 계산할 때, 각 유저별로 구해야할지 전체 유저별로 구해야할지...

사용자 유사도 계산 (코사인 유사도)

```python
matrix = df.iloc[:,4:-1].T
matrix
```

```python
# Compute the Cosine Similarity matrix based on the count_matrix
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(matrix, matrix)
cosine_sim
```

```python
idx = ['user17','user18','user21','user22','user23','user24','user25','user26','user27','user28']
cosine_sim_df = pd.DataFrame(cosine_sim, columns = idx)
cosine_sim_df #(10*10)
```

경로 예측 알고리즘은 구한 사용자 유사도 행렬과 weight를 곱한다.
그리고 ui의 tgpk에 대한 확률을 더 해준다.=>각 패턴별로 그때그때 사용자 합 구함.

```python
url_df = df[df['current url']=='HOME']
url_df
```

```python
user_df = cosine_sim_df['user17']
user_df
```

```python
user_p = url_df['weight'].to_numpy().reshape([len(url_df['weight']),1])*user_df.to_numpy().reshape([10,1]).T
user_p_sum = user_p.sum(axis=1)
user_p_sum
```

```python
import numpy as np
ind = np.argmax(user_p.sum(axis=1))
ind
```

```python
print(url_df.loc[ind][:4], round(user_p_sum[ind],2))
```

```python
import numpy as np
def predict_action(userid, current_url):
    url_df = df[df['current url']==current_url].reset_index(drop=True)
    user_df = cosine_sim_df[userid]
    # weight(패턴의 중요도 n*1) * cosine_sim_df(한 사람과 다른 사람들의 유사도 1*10) => (n*10)
    user_p = url_df['weight'].to_numpy().reshape([len(url_df['weight']),1])*user_df.to_numpy().reshape([10,1]).T
    user_p_sum = user_p.sum(axis=1)
    
    # max probability path
    idx = np.argmax(user_p_sum)
    return url_df.loc[idx][:4], round(user_p_sum[idx]*100,2)
```

```python
pred_path, pred_prob =  predict_action('user22','photo')
print(pred_prob, '%의 확률로')
print(pred_path)
```
